```python
import numpy as np
import glob
import os
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Flatten, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
```

```python
gesture = {
    0:'A', 1:"B", 2:"C", 3:'D', 4:'E', 5:'F', 6:'G', 7:'H', 8:'I', 9:'J', 10:'K', 11:'L', 12:'M', 13:'N', 14:'O', 15:'P',
    16:'Q', 17:'R', 18:'S', 19:'T', 20:'U', 21:'V', 22:'W', 23:'X', 24:'Y', 25:'Z',
}
```

```python
len(gesture)
```

```python
path = "C://ai_project01/hand_detect01/hand_dataset/*.npy"
```

```python
file_list = glob.glob(path)

file_list
```

```python
for file in file_list:
    print("file=",file)
    data = np.load(file)
    print("data.shape=",data.shape)
```

```python
data_list = []

for file in file_list:
    print("file=",file)
    data = np.load(file)
    print("data.shape=",data[:550,:].shape)
    data_list.extend(data[:550,:])
```

```python
len(data_list)
```

```python
all_data = np.array(data_list, dtype=np.float32)

all_data
```

```python
all_data.shape
```

```python
all_data_lstm = all_data.reshape(-1, 5, 79)
all_data_lstm.shape
```

```python
all_data_lstm
```

```python
X_data = all_data_lstm[:, :, :-1]
X_data
```

```python
labels = all_data_lstm[:, 0, -1]

labels
```

```python
y_data = to_categorical(labels, num_classes=len(gesture))

y_data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.1)
```

```python
X_train
```

```python
y_train[0:5]
```

```python
model = Sequential()
```

```python
model.add(LSTM(256, input_shape=(5,78)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(128, activation='relu'))
```

```python
model.add(Dropout(0.5) )
```

```python
model.add(Dense(64, activation='relu'))
```

```python
model.add(Dropout(0.5))
```

```python
model.add(Dense(32, activation='relu'))
```

```python
model.add(Dropout(0.5))
```

```python
model.add(Dense(len(gesture), activation="softmax"))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
```

```python
cb_checkpoint = ModelCheckpoint(filepath="c://ai_project01/hand_lstm_train_result/",
                                monitor="val_acc",
                                save_best_only=True
                               )
```

```python
earlystopping = EarlyStopping(monitor="val_loss",
                              patience=500,
                             )
```

```python
model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=1000,
    callbacks=[cb_checkpoint, earlystopping]
)
```
